```python
#Importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Atenas, Grecia.csv')
df
```

```python
df.info()
```

**a) Filtro 1: Registro de 5 hostname diferentes**

```python
filtro1 = df.drop_duplicates(subset='host_name').head(5)
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro_A(5 'host_name' diferentes utilizando 'drop duplicates').csv")
```

**b) Filtro 2: Registros de los host que se hallan unido a Airbnb antes del año 2019**

```python
filtro2 = df[df["host_since"] < "2019-01-01"]
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro_B('host_since' desde 2019).csv")
```

**c) Filtro 3: Los registros de los host que responden a más tardar en 1 dia**

```python
filtro3 = df[df["host_response_time"] == "within a day"]
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("Filtro_C('host_response_time' igual a 'within a day').csv")
```

**d) Filtro 4: Registros que tengan al menos 5 accommodates**

```python
filtro4 = df[df["accommodates"] >= 5]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv("Filtro_D('accommodates' mayor o igual a 5).csv")
```

**e) Filtro 5: Los registros de los tipos de cuarto: Entire home y Private room** 

```python
filtro5 = df[(df["room_type"] == "Entire home/apt") | (df["room_type"] == "Private room")]
filtro5
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv("Filtro_E('room_type' igual a 'Entire home' o 'Private room').csv")
```

**f) Filtro 6: Registros que a lo mucho cuenten con 4 camas**

```python
filtro6 = df[df["beds"] <= 4]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("Filtro_F('beds' menor o igual a 4).csv")
```

**g) Filtro 7: Registros que tengan 1 baño o que sean superhost, pero con la condición de que todos sean instant_bookable**

```python
filtro7 = df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"] == "t")]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("Filtro_G('bathrooms' igual a 1 o 'host_is_superhost' igual a t, pero 'instant_bookable' igual a t).csv")
```

**h) Filtro 8: Registros cuyo Review_scores_cleanliness sean menor de 3 y Review_scores_rating sea mayor de 3**

```python
filtro8 = df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro_H('review_scores_cleanliness' menor a 3 y 'review_scores_rating' mayor a 3).csv")
```
