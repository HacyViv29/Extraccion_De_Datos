```python
#Importamos librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
df.info()
```

**FILTRO 1: FILTRO POR OBJETO**

```python
#Filtro por objeto
filtro1=df[(df["NOMBRE_CLIENTE"]=="LUIS ENRIQUE PATIÑO PEREZ")]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1_Objeto(Nombre_cliente).csv")
```

**FILTRO 2: FILTRO POR FILAS**

```python
#Filtro por filas
filtro2=df.iloc[3:30, :] #Filas consecutivas
#filtro2=df.iloc[[2,5], :] #Filas no consecutivas
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro2_Filas(Fila 3-30).csv")
```

**FILTRO 3: FILTRO POR COLUMNAS**

```python
#Filtro por columnas
#filtro3=df.iloc[:, 1:3] #Columnas consecutivas
filtro3=df.iloc[:, [1,2,3,4,13]] #Columnas no consecutivas
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("Filtro3_Columnas(Fecha_Doc,Nombre_Vendedor,Nombre_Cliente,Cant,Margen_Total_Calculado).csv")
```

**Filtro 7: Filtro por comparación**

**Filtro 8: Filtro Y**

```python
#Filtro Y
filtro8=df[(df["CANT"] <= 3) & (df["MARGEN_TOTAL_CALCULADO"] > 1000)]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro8_FiltroY(CANT y MARGEN_TOTAL_CALCULADO).csv")
```

**Filtro 9: Filtro O**

```python
#Filtro O
filtro9=df[(df["NOMBRE_VENDEDOR"] == "DIEGO ARMANDO LEON DURAN") | (df["CANT"] >= 10)]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Filtro9_FiltroO(NOMBRE_VENDEDOR o CANT).csv")
```

**Filtro 10: Filtro Not**
